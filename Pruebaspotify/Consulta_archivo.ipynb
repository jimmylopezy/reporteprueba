```python
import pandas as pd
import json
```

```python
# Cargar el archivo JSON
with open('D:/Profesion/taylor_swift_spotify.json') as file:
    data = json.load(file)
```

```python
# Normalizar los datos JSON y convertirlo en DataFrame
df = pd.json_normalize(data, record_path=['albums', 'tracks'], 
                      meta=['artist_id', 'artist_name', 'artist_popularity', 
                            ['albums', 'album_id'], ['albums', 'album_name'], ['albums', 'album_release_date'], ['albums', 'album_total_tracks']])
```

```python
# Guardar como archivo CSV
df.to_csv('data.csv', index=False)
```

```python
df_spotify=df
```
